# Covid19 Data Analysis

# Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Read the covid19 dataset

```python
corona_dataset_csv = pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset_csv.head()
```

#### Shape of the dataframe

```python
corona_dataset_csv.shape
```

# Drop irrelevant attributes

```python
df = corona_dataset_csv.drop(columns = ['Lat', 'Long'])
df
```

```python
corona_dataset_aggregated = df.groupby('Country/Region').sum()
corona_dataset_aggregated
```

```python
corona_dataset_aggregated.shape
```

### Aggregating the observations by country

```python
corona_dataset_aggregated.loc['China'].plot()
corona_dataset_aggregated.loc['India'].plot()
corona_dataset_aggregated.loc['Italy'].plot()
plt.legend()
```

### Visualizing data related to a country for example China

```python
corona_dataset_aggregated.loc['China'][:3].plot()
```

```python
corona_dataset_aggregated.loc['China'].diff() # calculate first derivative 
```

```python
corona_dataset_aggregated.loc['China'].diff().plot()
```

# Calculating a good measure 
    we need to find a good measure reperestend as a number, 
    describing the spread of the virus in a country. 

```python
corona_dataset_aggregated.loc['China'].diff().max() # which means in only 1 day 15136 cases were reported
```

```python
corona_dataset_aggregated.loc['Italy'].diff().max()
```

```python
corona_dataset_aggregated.loc['India'].diff().max()
```

### Infection rate of countries

```python
def func(df):
    df['max diff'] = df.diff().max()
    return df
corona_data = corona_dataset_aggregated.apply(lambda x: func(x), axis=1)[['max diff']]
corona_data
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection_rates
```

# Importing worldwide happiness report

```python
happiness_report_csv = pd.read_csv('worldwide_happiness_report.csv')
happiness_report_csv.head()
```

```python
useless_cols = ['Overall rank', 'Score', 'Generosity', 'Perceptions of corruption']
```

```python
happiness_report_csv = happiness_report_csv.drop(columns = useless_cols)
```

```python
happiness_report_csv.set_index('Country or region', inplace=True)
```

```python
happiness_report_csv.head() 
```

```python
data=corona_data.join(happiness_report_csv, how='inner')
data
```
